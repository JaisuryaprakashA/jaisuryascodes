```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/spam_ham_dataset.csv")
```

```python
df.head()
```

```python
# prompt: generate code for  newly entered text is spam or ham

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Drop the 'Unnamed: 0' column if it exists, as it seems irrelevant based on typical CSV reads
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

# Map 'ham' to 0 and 'spam' to 1
df['label_num'] = df['label'].apply(lambda x: 1 if x == 'spam' else 0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label_num'], test_size=0.2, random_state=42)

# Vectorize the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000) # You can adjust max_features
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Train a Multinomial Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# Function to predict if a new text is spam or ham
def predict_spam_ham(text):
  """
  Predicts whether a given text is spam or ham using the trained model.

  Args:
    text: The input string to classify.

  Returns:
    'spam' or 'ham'.
  """
  text_tfidf = tfidf_vectorizer.transform([text])
  prediction = model.predict(text_tfidf)
  return 'spam' if prediction[0] == 1 else 'ham'

# Example usage:
new_text = "Congratulations! You've won a free iPhone. Click here to claim."
prediction = predict_spam_ham(new_text)
print(f"The text '{new_text}' is classified as: {prediction}")

new_text_2 = "Hey, just checking in. How are you doing?"
prediction_2 = predict_spam_ham(new_text_2)
print(f"The text '{new_text_2}' is classified as: {prediction_2}")

# Evaluate the model on the test set (optional)
y_pred = model.predict(X_test_tfidf)
print("\nModel Evaluation:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
# prompt: get input from user

user_input = input("Enter a message to classify: ")
prediction = predict_spam_ham(user_input)
print(f"The message is classified as: {prediction}")
```
